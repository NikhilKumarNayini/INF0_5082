```python
feature_columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
predicted_column=['Outcome']
x=df[feature_columns].values
y=df[predicted_column].values
```

# Train,Test and Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=69)
from sklearn.impute import SimpleImputer
fill_values = SimpleImputer(missing_values=0, strategy='mean')
X_train=fill_values.fit_transform(X_train)
X_test=fill_values.fit_transform(X_test)
```

# Random Forest Algorithm and confusion matrix

```python
random_forest_model=RandomForestClassifier(random_state=69)
random_forest_model.fit(X_train,y_train.ravel())
predict_train_data=random_forest_model.predict(X_test)
rand_forest_Accuracy=mc.accuracy_score(y_test, predict_train_data)
print(rand_forest_Accuracy)
cm = mc.confusion_matrix(y_test, predict_train_data)
sns.heatmap(cm, cmap="RdYlGn",fmt="d",annot =True, yticklabels = ['Healthy', 'Diabetic'], xticklabels = ['Predicted Healthy', 'Predicted Diabetic'])
plt.show()
```

# Logistic Regression and confusion Matrix

```python
log_reg_model = LogisticRegression(C= 0.5, random_state= 69)
log_reg_model.fit(X_train, y_train.ravel())
y_pred = log_reg_model.predict(X_test)
y_prob = log_reg_model.predict_proba(X_test)[:,1]
log_reg_Accuracy=metrics.accuracy_score(y_test, y_pred)
print(log_reg_Accuracy)
cm=mc.confusion_matrix(y_test,y_pred)
sns.heatmap(cm, cmap='RdYlGn',fmt="d",annot=True, yticklabels=['Healthy','Diabetic'],xticklabels=['Predicted Healthy','Predicted Diabetic'])
plt.show()
```

# Support Vector Classification and Confusion Matrix

```python
#Fitting SVC Model
from sklearn.svm import SVC
from sklearn import metrics
SVC_classifier = SVC(C= 0.7, kernel= 'linear', random_state= 69, probability=True)
SVC_classifier.fit(X_train, y_train)
y_pred = SVC_classifier.predict(X_test)
y_prob = SVC_classifier.predict_proba(X_test)[:,1]
SVC_Accuracy=metrics.accuracy_score(y_test, y_pred)
print(SVC_Accuracy)
cm = mc.confusion_matrix(y_test, y_pred)
sns.heatmap(cm, cmap='RdYlGn',fmt="d",annot=True, yticklabels=['Healthy','Diabetic'],xticklabels=['Predicted Healthy','Predicted Diabetic'])
plt.show()
```

# K-Nearest Neighbors Classifier Model and its confusion Matrix

```python
#Fitting KNeighborsClassifier Model
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
classifier = KNeighborsClassifier(metric= 'manhattan', n_neighbors= 10)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_prob = classifier.predict_proba(X_test)[:,1]
KNeighbours_Accuracy=metrics.accuracy_score(y_test, y_pred)
print(KNeighbours_Accuracy)
cm = mc.confusion_matrix(y_test, y_pred)
sns.heatmap(cm, cmap='RdYlGn', fmt="d",annot=True, yticklabels=['Healthy','diabetic'], xticklabels =['Predicted_Healthy','Predicted_diabetic'])
plt.show()
```

### append accuraccies to an array

```python
Accuracy=[]
Accuracy.append(log_reg_Accuracy)
Accuracy.append(rand_forest_Accuracy)
Accuracy.append(SVC_Accuracy)
Accuracy.append(KNeighbours_Accuracy)
```

# Accuracy comparision using Bar Chart

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0,0,0.8,0.8])
Models = ['logisticRegression','RandomForestRegression','SVC','KNeighbors_Accuracy']
ax.bar(Models,Accuracy)
plt.title("Accuracy comparision of Machine Learning Models")
plt.ylabel("Accuracies")
plt.xlabel("Machine Learning Models")
plt.ylim((0.6,0.8))
plt.show()
```
